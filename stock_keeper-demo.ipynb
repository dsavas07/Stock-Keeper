# Stock Keeper Demo Notebook
This notebook runs the full end-to-end Stock Keeper agent: data ingestion, forecasting, scoring, optimization, and plotting results. Use a valid Finnhub API key.

```python
import os
import matplotlib.pyplot as plt
from dotenv import load_dotenv
from stock_keeper.interface.cli import recommend_portfolio, recommend_from_prompt
from stock_keeper.pipeline import PortfolioAgent
load_dotenv()
print('Stock Keeper version: ready')
```

## CLI-style Portfolio Recommendation

```python
api_key = os.getenv('FINNHUB_API_KEY')
if not api_key:
    print('Set FINNHUB_API_KEY environment variable and rerun this cell.')
else:
    symbols = ['AAPL', 'MSFT', 'GOOGL', 'NVDA']
    out = recommend_portfolio(symbols, api_key, horizon=60, risk_appetite='medium')
    print('Portfolio recommendation:')
    print(out)
```

## Question-based Prompt Recommendation

```python
prompt = 'For moderate risk and long-term growth in tech, recommend best mix for AAPL MSFT NVDA'
if not api_key:
    print('Set FINNHUB_API_KEY to run prompt-based recommendation.')
else:
    result = recommend_from_prompt(prompt, api_key)
    print('Prompt -> result:')
    print(result)
    labels = []
    allocs = []
    for t,w in result.get('allocation', {}).items():
        labels.append(t)
        allocs.append(w)
    if labels:
        fig, ax = plt.subplots(figsize=(8, 5))
        ax.pie(allocs, labels=labels, autopct='%1.1f%%', startangle=90)
        ax.set_title('Recommended Allocation')
        plt.show()
```

## Portfolio Metrics and Notes

```python
if api_key:
    agent = PortfolioAgent(api_key)
    run = agent.recommend(['AAPL', 'MSFT', 'NVDA'], horizon=90, risk_appetite='medium', sector_constraints={'Technology': 0.5})
    print('Metrics:', run.get('portfolio_metrics'))
    print('Risk notes:', run.get('risk_notes'))
else:
    print('Set FINNHUB_API_KEY to run full metrics.')
```
