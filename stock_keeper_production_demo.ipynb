# Stock Keeper Production Demo
Use this notebook to run the production-ready Stock Keeper pipeline on a sample market set.

```python
import os
from stock_keeper.interface.agent import InteractivePortfolioAgent
from dotenv import load_dotenv
import matplotlib.pyplot as plt
load_dotenv()
## print('Ready')
print(f'FINNHUB_API_KEY: {os.getenv('FINNHUB_API_KEY')}')
```

```python
api_key = os.getenv('FINNHUB_API_KEY')
if not api_key:
    raise ValueError('Set FINNHUB_API_KEY in environment variables before running this notebook.')
agent = InteractivePortfolioAgent(api_key)
question = 'Recommend a moderate-risk long-term one-month portfolio for AAPL MSFT NVDA and INTC.'
result = agent.answer_question(question, symbols=['AAPL','MSFT','NVDA','INTC'], risk_appetite='medium', horizon=90, sector_constraints={'Technology': 0.5})
result
```

```python
alloc = result.get('allocation', {})
symbols = list(alloc.keys())
weights = list(alloc.values())
plt.figure(figsize=(7, 5))
plt.pie(weights, labels=symbols, autopct='%1.1f%%')
plt.title('Recommended Portfolio Allocation')
plt.show()
print('Portfolio metrics:', result.get('portfolio_metrics'))
print('Risk notes:', result.get('risk_notes'))
```
