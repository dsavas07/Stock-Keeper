## Finlight POST Fetch Articles

```python
import os
from finlight_client import FinlightApi, ApiConfig
from finlight_client.models import GetArticlesParams
import pandas as pd
import json
import duckdb
from functools import reduce
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
from datetime import datetime, timedelta

# Get the current time
now = datetime.now()

# Subtract 24 hours
twenty_four_hours_ago = now - timedelta(hours=24)

def finlight_get_articles(api_key: str, query: str = '', sources: list = ['*'], 
    excludeSources: list = [], 
    countries: list = [], 
    categories: str = "markets,economy,business,politics,geopolitics,regulation,technology,energy,commodities,crypto,health,climate,security", 
    from_date: str = twenty_four_hours_ago.strftime("%Y-%m-%dT%H:%M"), 
    to_date: str = now.strftime("%Y-%m-%dT%H:%M"),
    pageSize: int = 1000, page: int = 1, 
    orderBy: str = 'publishDate', order: str = 'DESC'
) -> pd.DataFrame:
    client = FinlightApi(
        config=ApiConfig(
            api_key=api_key
        )
    )

    params = GetArticlesParams(query=query, 
    sources=sources, excludeSources=excludeSources, 
    countries=countries, categories=categories, 
    from_=from_date, to=to_date, 
    pageSize=pageSize, page=page, orderBy=orderBy, order=order)
    articles = client.articles.fetch_articles(params=params)

    articles_df = pd.DataFrame([article.model_dump() for article in articles.articles])
    return articles_df
```

```python
articles_df = finlight_get_articles(api_key = os.getenv('FINLIGHT_API_KEY'), query='ticker:AAPL')
articles_df.head()
```
