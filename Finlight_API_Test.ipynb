## Finlight POST Fetch Articles

```python
import os
from finlight_client import FinlightApi, ApiConfig
from finlight_client.models import GetArticlesParams
import pandas as pd
import json
import duckdb
from functools import reduce
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
from datetime import datetime, timedelta

finlight_api_key = os.getenv('FINLIGHT_API_KEY')

# Get the current time
now = datetime.now()

# Subtract 24 hours
twenty_four_hours_ago = now - timedelta(hours=24)

def finlight_get_articles( 
    query = '', 
    sources = ['*'], 
    excludeSources = ['economictimes.indiatimes.com'], 
    countries = ['US'], 
    ## categories: str | None = "markets,economy,business,politics,geopolitics,regulation,technology,energy,commodities,crypto,health,climate,security", 
    from_date = twenty_four_hours_ago.strftime("%Y-%m-%dT%H:%M"), 
    to_date = now.strftime("%Y-%m-%dT%H:%M"),
    language = 'en',
    includeEntities = True,
    excludeEmptyContent = True,
    includeContent = True,
    pageSize = 100, 
    page = 1, 
    orderBy = 'publishDate', order = 'DESC'
) -> pd.DataFrame:
    client = FinlightApi(
        config=ApiConfig(
            api_key=finlight_api_key
        )
    )

    params = GetArticlesParams(query=query, 
    sources=sources, 
    excludeSources=excludeSources, 
    countries=countries, 
    ## categories=categories, 
    from_=from_date, 
    to=to_date, 
    language=language,
    includeEntities=includeEntities,
    excludeEmptyContent=excludeEmptyContent,
    includeContent=includeContent,
    pageSize=pageSize, 
    page=page, 
    orderBy=orderBy, order=order
    ) 
    articles = client.articles.fetch_articles(params=params)

    articles_df = pd.DataFrame([article.model_dump() for article in articles.articles])
    return articles_df
```
